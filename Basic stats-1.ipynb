# Descriptive Analytics for Numerical Columns

```python
# To find the mean for numerical data
df.mean(numeric_only=True)
```

```python
#To find the median for numerical data
df.median(numeric_only=True)
```

```python
# To Find the mode for numerical data
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
```

```python
# To find the Standard Deviation for numerical data
df.std(numeric_only=True)
```

```python
# To Identify the numerical columns in the dataset
numerical_columns=['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']
```

```python
# To calculate basic statistics for numerical data
statistics={'Mean':df[numerical_columns].mean(),
            'Median':df[numerical_columns].median(),
            'Standard deviation':df[numerical_columns].std(),
            'Mode':df[numerical_columns].mode().iloc[0]}
```

```python
# Converting to a data frame for better readability
stats_df=pd.DataFrame(statistics)
print(stats_df)
```

# Brief Interpretation of these statistics

#Volume (Units Sold per Transaction) :
The average transaction involves 5.07 units, but the most common purchase size is 3 units.The median of 4 units suggests that many transactions are on the lower end.A standard deviation of 4.23 indicates moderate variability, meaning some larger purchases influence the average.

#Average Price per Unit ($) :
The average price is 10,453, but the median price is much lower at 1,450, suggesting skewed pricing with some expensive items.
The most frequently occurring price point is 400, indicating that lower-priced items dominate the sales.
A high standard deviation of 18,079 confirms significant price variations.

#Total Sales Value per Transaction ($) :
The average total sales value is 33,812, but the median is only 5,700, indicating that most transactions are much smaller. 
The mode of 24,300 suggests that some sales cluster around this value. 
A high standard deviation of 50,535 confirms that a few very large transactions are skewing the mean.

#Discount Rate (%) :
The average discount applied is 15.16%, with a median of 16.58%. 
The most common discount rate is 5.01%, meaning some products receive minimal discounts.
A standard deviation of 4.22% suggests relatively low variability in discount percentages.

#Net Sales Value (After Discounts) ($) :
The average net sales value is 30,466, but the median is just 4,677, suggesting that smaller transactions are more common.
The mode of 326 indicates that many low-value transactions occur.
A standard deviation of 46,358 highlights a wide range in net sales values, with some very high-value transactions pulling up the average.

# Data Visualization

# Histograms

```python
import matplotlib.pyplot as plt

for col in numerical_cols:
    df[col].hist(bins=20)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

# Analysis

# Box Plots:

```python
df[numerical_cols].plot(kind='box', subplots=True, layout=(1, len(numerical_cols)), figsize=(15,5))
plt.show()
```

# Analysis

# Bar Chart Analysis for Categorical Column:

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
