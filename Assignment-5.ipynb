```python
#checking for missing values
df.isna().sum()
```

```python
#handling missing values
df.fillna(df.median(),inplace=True)
df.isna().sum()
```

```python
#visualization
plt.figure(figsize = (6,5))
sns.boxplot(x=df["ALTV"])
plt.title('Boxplot of Abnormal Long Term Variability(ALTV)')
plt.xlabel('Abnormal Long Term Variability (ALTV)')
plt.show()         
```

```python
# calculate IQR
Q1 = df['ALTV'].quantile(0.25)
Q3 = df['ALTV'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Detect outliers using IQR
outliers_iqr = df[(df['ALTV'] < lower_bound) | (df['ALTV'] > upper_bound)]
print(f"Number of outliers detected using IQR: {outliers_iqr.shape[0]}")
```

```python
# Cap outliers
df['ALTV'] = df['ALTV'].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Visualizing after removing the outliersplt.figure(figsize=(5,4))
sns.boxplot(x=df["ALTV"])
plt.title('Boxplot of Abnormal Long Term Variability(ALTV)')
plt.xlabel('Abnormal Long Term Variability (ALTV)')
plt.show()
```

```python
#summary
# Initialize a dictionary to store the statistics
summary_stats = {}

# Iterate through each column in the DataFrame
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        # Calculate mean, median, standard deviation, and IQR
        mean = df[column].mean()
        median = df[column].median()
        std_dev = df[column].std()
        Q1 = df[column].quantile(0.25)
        Q3 =df[column].quantile(0.75)
        IQR = Q3 - Q1
        
        # Store the results in the dictionary
        summary_stats[column] = {
            'Mean': mean,
            'Median': median,
            'Standard Deviation': std_dev,
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR
        }
# Convert the dictionary to a DataFrame for easier viewing
summary_df = pd.DataFrame(summary_stats).T

# Display the summary statistics
print(summary_df)
```
